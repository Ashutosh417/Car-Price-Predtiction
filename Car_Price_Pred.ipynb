```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['country'] = le.fit_transform(df['country'])
```

```python
df['car purchase amount in tenthousand'] = df['car purchase amount']/10000
```

```python
df.drop(columns=['car purchase amount'],inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['car purchase amount in tenthousand']),
                                                         df['car purchase amount in tenthousand'],
                                                         test_size=0.2)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms = MinMaxScaler()
X_train_trans = ms.fit_transform(X_train)
X_test_trans = ms.transform(X_test)
```

```python
ms1 = MinMaxScaler()
y_train_trans = ms1.fit_transform(y_train.values.reshape(-1,1))
y_test_trans = ms1.transform(y_test.values.reshape(-1,1))
```

```python
X_train_trans
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(64, activation='relu',input_dim=6))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='linear'))

model.summary()
          



```

```python
model.compile(optimizer='Adam',loss='mean_squared_error')
```

```python
history = model.fit(X_train_trans,y_train_trans,validation_split=0.1,epochs=500)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    score = r2_score(actual, pred)
    return print("r2_score:", score, "\nmae:", mae, "\nmse:", mse, "\nrmse:", rmse)
```

```python
y_pred = model.predict(X_test_trans)
```

```python
eval_metric(y_test_trans,y_pred)
```

```python
y_pred_inv = ms1.inverse_transform(y_pred)
y_test_inv = ms1.inverse_transform(y_test_trans)
```

```python
eval_metric(y_test_inv,y_pred_inv)
```

```python
res = (y_pred_inv)-(y_test_inv)
```

```python
%matplotlib inline
plt.figure(figsize=(8,5))
plt.scatter(y_pred,res)
plt.axhline(y=0,linewidth=3,color='green')
plt.xlabel('Predicted_Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted value')
```
